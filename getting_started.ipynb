# Getting Started with the Utah Housing Dataset

ACS 5-year estimates for Utah census tracts, 2009–2023.

Run this notebook to fetch the data, save it locally, and do an initial exploration. After saving to `data/utah_housing.csv`, you can reload the CSV directly without calling the Census API again.

For a full written walkthrough of each step, see the [Tutorial page](https://congerm.github.io/Utah_Housing_Project/Tutorial.html) on the project website.

## 1. Setup

```python
import os
import pandas as pd
from utah_housing import fetch_all_years, OUTCOME, PREDICTORS
```

## 3. Data Acquisition

Fetch ACS 5-year estimates for all Utah census tracts across 2009–2023.

```python
df = fetch_all_years(years=range(2009, 2024))
print(f"Total rows: {len(df):,}")
df.head()
```

## 4. Save Data

```python
os.makedirs("data", exist_ok=True)
df.to_csv("data/utah_housing.csv", index=False)
print("Saved to data/utah_housing.csv")
```
